### Проверьте, что эффективность обогащения рассчитана правильно. Вычислите её на обучающей выборке для признака rougher.output.recovery. Найдите MAE между вашими расчётами и значением признака. Опишите выводы.

```python
def recovery(c, f, t):
    r = ( (c * (f - t)) / (f * (c - t)) ) * 100
    return r
```

## Модель

Т.к. целевой признак количественный, решаем задачу регрессии.

### Напишите функцию для вычисления итоговой sMAPE.

```python
def smape(target, predictions):
    
    r = sum ( (((abs(target[target.columns[0]] - predictions[:,0])) / ((abs(target[target.columns[0]]) + abs(predictions[:,0])) / 2)) * 100) )
    
    r_err = r / target.shape[0]
     
    f = sum ( (((abs(target[target.columns[1]] - predictions[:,1])) / ((abs(target[target.columns[1]]) + abs(predictions[:,1])) / 2)) * 100) )
    
    f_err = f / target.shape[0]

    err = 0.25 * r_err + 0.75 * f_err
    
    return err
```

### Обучите разные модели и оцените их качество кросс-валидацией. Выберите лучшую модель и проверьте её на тестовой выборке. Опишите выводы.

```python
print('Количество столбцов features_train =', len(features_train.columns),
      '\nКоличество столбцов data_test =', len(data_test.columns)
     )
display('Признаки data_test:', data_test.columns,
        'Признаки features_train:', features_train.columns
       )
```

```python
scorer = make_scorer(smape, greater_is_better=False)
```

```python
model0 = LinearRegression()
model0.fit(features_train, target_train)
predict0 = model0.predict(data_test)
```

```python
cvs0 = cross_val_score(model0, features_train, target_train, scoring=scorer) 

display('Качество кросс-валидации модели 0:', cvs0)
print('Средняя оценка качества модели 0 =', cvs0.mean())
```

```python
final_smape0 = smape(target_test, predict0)

print('Итоговое sMAPE модели 0 =', final_smape0)
```

```python
depth = [int(x) for x in np.linspace(1, 51, num=5)]
split = [2, 4, 6]
leaf = [1, 2, 4]

grid = {'max_depth': depth, 
         'min_samples_split': split, 
         'min_samples_leaf': leaf
        }
```

```python
model1 = GridSearchCV(DecisionTreeRegressor(), grid, scoring=scorer, cv=5)

model1.fit(features_train, target_train)

best_params_m1 = model1.best_params_

display('Наилучшие гиперпараметры для модели 1:', best_params_m1)

predict1 = model1.predict(data_test)
```

```python
cvs1 = cross_val_score(model1, features_train, target_train, scoring=scorer) 

display('Качество кросс-валидации модели 1:', cvs1)
print('Средняя оценка качества модели 1 =', cvs1.mean())
```

```python
final_smape1 = smape(target_test, predict1)

print('Итоговое sMAPE модели 1 =', final_smape1)
```

```python
model2 = RandomForestRegressor()
model2.fit(features_train, target_train)
predict2 = model2.predict(data_test)
```

```python
cvs2 = cross_val_score(model2, features_train, target_train, scoring=scorer) 

display('Качество кросс-валидации модели 2:', cvs2)
print('Средняя оценка качества модели 2 =', cvs2.mean())
```

```python
final_smape2 = smape(target_test, predict2)

print('Итоговое sMAPE модели 2 =', final_smape2)
```

```python
if final_smape0 < final_smape1:
    best = final_smape0
    best_model = model0
else:
    best = final_smape1
    best_model = model1

if best > final_smape2:
    best = final_smape2
    best_model = model2
    
print('Наименьшее sMAPE =', best, 'принадлежит модели', best_model)
```

Проверим модель с наименьшим sMAPE на тестовой выборке.

```python
predict_test = best_model.predict(data_test)

display('Лучшая модель: Предсказания эффективности обогащения чернового концентрата rougher.output.recovery и финального концентрата final.output.recovery', predict_test)

final_smape = smape(target_test, predict_test)

print('Лучшая модель: Итоговое sMAPE =', final_smape)
```

Рассмотрим константную модель для оценки адекватности предсказаний.

```python
model_const = DummyRegressor()
model_const.fit(features_train, target_train)
predict_const = model_const.predict(data_test)
```

```python
final_smape_c = smape(target_test, predict_const)

print('Итоговое sMAPE константной модели =', final_smape_c)
```

Выбранная модель является адекватной.
